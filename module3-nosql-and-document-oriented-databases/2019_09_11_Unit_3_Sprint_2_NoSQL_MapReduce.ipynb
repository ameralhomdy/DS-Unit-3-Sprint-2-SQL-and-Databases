### Load RPG from SQLite

````python
import sqlite3

try:
    sqlte_conn = sqlite3.connect('rpg_db.sqlite3')
    sqlte_curs = sqlte_conn.cursor()
    # Load RPG Game player characters from the Character table
    characters = sqlte_curs.execute('SELECT * FROM charactercreator_character;').fetchall()
    for character in characters[:5]:
        print(character)
except Exception as e:
    print(e)
````

````python
id = characters[0][0]
name = characters[0][1]
level = characters[0][2]
xp = characters[0][3]
print(f'id: {id}, name: {name}, level: {level}, xp: {xp}')
````

### Create RPG database in MongoDB

````python
rpg_db = client.rpg
rpg_db
````

### Insert RPG characters from Python List to MongoDB `rpg_db`

````python
for character in characters:
    rpg_db.rpg.insert_one({
        'Character ID' : character[0],
        'Name' : character[1],
        'Level' : character[2],
        'EXP' : character[3],
        'HP' : character[4],
        'Strength' : character[5],
        'Intelligence' : character[6],
        'Dexterity' : character[7],
        'Wisdom' : character[8]
    })
    
list(rpg_db.rpg.find()) 
````
